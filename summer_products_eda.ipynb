```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv("/content/summer-products-with-rating-and-performance_2020-08.csv")
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

```python
df.columns
```

```python
custom_palette = sns.color_palette("Paired")
sns.set_palette(custom_palette)
```

```python
import plotly.express as px

fig_product_color = px.bar(df['product_color'].value_counts().reset_index(),
                            x='index', y='product_color',
                            labels={'index': 'Product Color', 'product_color': 'Count'},
                            title='Product Color Distribution')
fig_product_color.update_xaxes(tickangle=45)
fig_product_color.show()
```

```python
qfig_product_rating = px.histogram(df, x='rating', nbins=10,
                                  labels={'rating': 'Rating'},
                                  title='Product Rating Distribution')

fig_product_rating.show()
```

```python
fig_origin_country = px.pie(df, names='origin_country', title='Origin Country Distribution')
fig_origin_country.show()
```

```python
fig_currency_buyer = px.pie(df, names='currency_buyer', title='Currency Buyer Distribution')
fig_currency_buyer.show()
```

```python
fig_scatter = px.scatter(df, x='merchant_rating_count', y='merchant_rating', color='merchant_rating',
                         title='Merchant Rating Count vs. Merchant Rating')
fig_scatter.update_traces(marker=dict(size=8, opacity=0.7))

fig_scatter.update_layout(
    xaxis_title='Merchant Rating Count',
    yaxis_title='Merchant Rating',
    legend_title='Merchant Rating',
    font=dict(size=12),
)

fig_scatter.show()
```

```python
fig_scatter_rating = px.scatter(df, x='rating_count', y='rating',
                                title='Rating vs. Rating Count',
                                labels={'rating_count': 'Rating Count', 'rating': 'Rating'})

fig_scatter_rating.show()
```

```python
fig_pie = px.pie(df, names='uses_ad_boosts', title='Ad Boosts Usage Distribution',
                 color_discrete_map={0: 'red', 1: 'green'})

fig_pie.show()
```

```python
fig_scatter = px.scatter(df, x='units_sold', y='rating', color='uses_ad_boosts',
                         title='Units Sold vs. Rating with Ad Boosts',
                         labels={'units_sold': 'Units Sold', 'rating': 'Rating'},
                         color_discrete_map={0: 'red', 1: 'green'})

fig_scatter.show()
```

```python
fig_urgency_units = px.bar(df, x='urgency_text', y='units_sold',
                           title='Units Sold based on Urgency Text',
                           labels={'urgency_text': 'Urgency Text', 'units_sold': 'Units Sold'})

fig_urgency_units.show()
```
