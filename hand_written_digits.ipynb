```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python

 (X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
import random
def show_img(data, label_names):
    fig, axs = plt.subplots(2, 2, figsize=(8, 8))

    for i in range(2):
        for j in range(2):
            rand = random.randint(0, len(data) - 1)
            axs[i, j].imshow(data[rand])
            axs[i, j].set_title(label_names[rand])
            axs[i, j].axis('off')

    plt.show()
```

```python
show_img(X_train, y_train)
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
X_train_norm = X_train / 255.0
X_test_norm = X_test / 255.0
```

```python
tf.random.set_seed(42)

model_1 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(4, activation="relu"),
    tf.keras.layers.Dense(4, activation="relu"),
    tf.keras.layers.Dense(4, activation="relu"),
    tf.keras.layers.Dense(10, activation="softmax")
])

model_1.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                optimizer=tf.keras.optimizers.Adam(),
                metrics=["accuracy"])

history_1 = model_1.fit(X_train_norm,
                      y_train,
                      epochs=10,
                      validation_data=(X_test_norm,y_test))
```

```python
pd.DataFrame(history_1.history).plot()
```

```python
predictions = model_1.predict(X_test_norm)
predicted_labels = tf.argmax(predictions, axis=1)
```

```python
predicted_labels
```

```python
def show_data_with_labels(X_test, model, y_test):
    random_images = np.random.randint(0, len(X_train), size=64)
    fig, axes = plt.subplots(nrows=8, ncols=8, figsize=(8, 8))

    predictions = model.predict(X_test[random_images])
    predicted_labels = np.argmax(predictions, axis=1)

    for i, ax in enumerate(axes.flatten()):
        image = X_test[random_images[i]]
        image = image.reshape(28, 28)
        ax.imshow(image, cmap='gray')
        ax.axis('off')
        true_label = y_test[random_images[i]]
        predicted_label = predicted_labels[i]

        if predicted_label == true_label:
            color = "green"
        else:
            color = "red"

        ax.set_title(f'True: {true_label}\nPred: {predicted_label}', color=color)

    fig.suptitle("Label, Prediction", fontsize=14)
    plt.tight_layout()
    plt.show()
```

```python
show_data_with_labels(X_test, model_1, y_test)
```
